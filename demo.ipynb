### 5. Grammatical assumptions
___

```python
# tokens and grammar
letter_tokens = TokenizerTransducer.read_json("Tokens/letters.json")                # letters for natural language process
tex_tokens = TokenizerTransducer.read_json("Tokens/tex_letters.json")               # words of the TeX language
grammar = GrammarAcceptor.read_from_dict(tex_tokens, "Grammar/tex_grammar.json")    # base grammar

# text normalization pipeline (french)
r1 = RulesTransducer(letter_tokens, letter_tokens, complete=False, drop=False, text_file='Rules/fr/Normalizer/toMin.txt')                 # normalization
r2 = RulesTransducer(letter_tokens, letter_tokens, complete=False, drop=False, text_file='Rules/fr/Normalizer/spacer.txt')                # normalization
r3 = RulesTransducer(letter_tokens, letter_tokens, complete=False, drop=False, text_file='Rules/fr/Normalizer/supp.txt')                  # normalization
r4 = RulesTransducer(letter_tokens, letter_tokens, complete=True , drop=False, text_file='Rules/fr/Lemmatizer/lemm.txt')                  # lemmatization

# lexical transduction (french)
r5 = RulesTransducer(letter_tokens, tex_tokens, complete=True , drop=True, text_file='Rules/fr/SeqToTex/math_rules.txt')                  # lexical transduction

# grammar pipeline
r6 = ChainTypeTransducer(grammar, name_type='dig', open_symbol='##<', close_symbol='>##')                     # numbers recognition
rg_list = GrammarTransducer.layerComposerSeries(grammar, 'Grammar/GRules')                                    # grammar composer
```

### 8. Statistics
___

```python
from Datasets.Audiodataset import *
```

```python
dataset = Audiodata()
```

```python
durations_dna = []
durations_nl = []

word_by_sec_dna = []
word_by_sec_nl = []

total_dna = 0.
total_nl = 0.

for i in range(len(dataset)):

    sample = dataset[i]
    audio = sample['audio']
    audio, sample_rate = torchaudio.load(audio)
    text = sample["sequence"]
    word_length = len(text.split())

    if audio.dim() == 2 : audio = torch.mean(audio, dim=0) # stereo
    elif audio.dim() != 1 : raise Exception("Error with audio dimension")

    duration = audio.size()[0]/sample_rate

    if sample['dna'] :
        durations_dna.append(duration)
        total_dna += duration
    else : 
        durations_nl.append(duration)
        total_nl += duration
        

print("  * Audio mean +- std duration")
print("      (C) ", np.mean(durations_dna), '+-', np.std(durations_dna))
print("     (NL) ", np.mean(durations_nl), '+-', np.std(durations_nl))
print("  * Audio total duration")
print("      (C) ", total_dna)
print("     (NL) ", total_nl)
```

### 9. Transducers
___

```python
from RulesTransduction.tokenizer import *
from RulesTransduction.rulesFST import * 
from RulesTransduction.grammarFST import * 

from Seq2Tex import * 
```

```python
seq_letters = list('abcdefghijklmnopqrstuvwxyz')
tex_letters = ['+', '=', '\\alpha', '\\cos', '\\sum']
```

```python
sigma_seq = TokenizerTransducer(seq_letters, add_space=True)
sigma_tex = TokenizerTransducer(tex_letters, add_space=False)
```

```python
r1 = pynini.cross(pynini.accep('p l u s', token_type=sigma_seq.table()), pynini.accep('+', token_type=sigma_tex.table())).set_input_symbols(sigma_seq.table()).set_output_symbols(sigma_tex.table())
r2 = pynini.cross(pynini.accep('e q u a l', token_type=sigma_seq.table()), pynini.accep('=', token_type=sigma_tex.table())).set_input_symbols(sigma_seq.table()).set_output_symbols(sigma_tex.table())
r3 = pynini.cross(pynini.accep('a l p h a', token_type=sigma_seq.table()), pynini.accep('\\alpha', token_type=sigma_tex.table())).set_input_symbols(sigma_seq.table()).set_output_symbols(sigma_tex.table())
r4 = pynini.cross(pynini.accep('c o s i n u s', token_type=sigma_seq.table()), pynini.accep('\\cos', token_type=sigma_tex.table())).set_input_symbols(sigma_seq.table()).set_output_symbols(sigma_tex.table())
r5 = pynini.cross(pynini.accep('s u m', token_type=sigma_seq.table()), pynini.accep('\\sum', token_type=sigma_tex.table())).set_input_symbols(sigma_seq.table()).set_output_symbols(sigma_tex.table())
```

```python
pynini.union(r1, r2, r3, r4, r5).optimize()
```
